```python
# Daily Reflection Agent - Deterministic Decision Tree (No LLM Hallucination)
# Guardrails: No free-form generation, no psychological diagnosis, fixed outputs.

def run_reflection_agent():
    print(" Daily Reflection Agent")
    print("Answer honestly – I will ask only predefined questions.\n")

    # Q1
    answer1 = input("Did you complete your main goal today? (yes/no): ").strip().lower()

    if answer1 == "yes":
        # Q2 – YES branch
        mood = input("How did you feel? (productive/neutral/stressed): ").strip().lower()
        
        if mood == "productive":
            print("\n Insight: You turn intentions into action.")
            print("Action: Keep a “win log” daily to build momentum.")
        elif mood == "neutral":
            print("\n Insight: Completion without satisfaction often means low alignment.")
            print("Action: Tomorrow, rate goal importance (1–5) before starting.")
        elif mood == "stressed":
            print("\n Insight: You achieved the goal but depleted energy.")
            print(" Action: Break next goal into 3 smaller steps + schedule a break.")
        else:
            print("\n Please choose from: productive, neutral, stressed.")
    
    elif answer1 == "no":
        # Q2 – NO branch
        reason = input("Why not? (lack of time / distraction / low motivation / unclear goal): ").strip().lower()
        
        if reason == "lack of time":
            print("\n Insight: Time may be leaking due to poor estimation or interruptions.")
            print(" Action: Track time in 30-min blocks for one day.")
        
        elif reason == "distraction":
            # Q3 – Distraction subtype
            dist_type = input("Type of distraction? (phone / social media / environment): ").strip().lower()
            if dist_type == "phone":
                print("\n Insight: Your phone becomes a high-frequency interruptor.")
                print(" Action: Use “focus mode” 25 min before touching your main goal.")
            elif dist_type == "social media":
                print("\n Insight: Social media pulls you into reactive consumption.")
                print(" Action: Log out of accounts until goal is done.")
            elif dist_type == "environment":
                print("\n Insight: Your environment is not optimized for deep work.")
                print(" Action: Create a 2-hour “do not disturb” physical setup.")
            else:
                print("\n Please choose: phone, social media, or environment.")
        
        elif reason == "low motivation":
            # Q3 – Low motivation subtype
            mot_type = input("Why low motivation? (burnout / no clarity / task too hard): ").strip().lower()
            if mot_type == "burnout":
                print("\n Insight: You might be overloading without recovery.")
                print(" Action: Take a real 30-min rest before deciding next task.")
            elif mot_type == "no clarity":
                print("\n Insight: Vague goals kill drive.")
                print(" Action: Rewrite goal as “I will [specific action] at [specific time]”.")
            elif mot_type == "task too hard":
                print("\n Insight: Difficulty without scaffolding leads to avoidance.")
                print(" Action: Reduce scope to 5% version (e.g., write 1 sentence, open 1 file).")
            else:
                print("\n Please choose: burnout, no clarity, or task too hard.")
        
        elif reason == "unclear goal":
            print("\n Insight: A fuzzy goal guarantees uncertain execution.")
            print(" Action: Tomorrow: define “done” in one measurable sentence before starting.")
        
        else:
            print("\n Please choose: lack of time, distraction, low motivation, or unclear goal.")
    
    else:
        print("\n Please answer 'yes' or 'no'.")

    print("\n Guardrail reminder: No psychological diagnosis given. This is a deterministic reflection tool.")

if __name__ == "__main__":
    run_reflection_agent()
```
